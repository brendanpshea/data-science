## How Do We Choose the Right Analytical Approach?

Before diving into specific analysis techniques, it's crucial to establish a systematic process for determining the most appropriate type of analysis for a given business problem. This process ensures that we're asking the right questions, using the correct data, and applying the most suitable analytical methods. Let's break down this process into three key steps:

### 1. Review and Refine Business Questions

The first step in any data analysis project is to clearly define the business questions you're trying to answer. This involves:

- Work closely with **stakeholders** like business leaders, managers, and other relevant parties to understand their needs and objectives.
- Clarify ambiguities by ensuring that all terms are well-defined and that the sc Identify which questions are most critical to the business's success.
- Translate business questions into specific, measurable analytical questions.

For example, at Juicy Lucy's, a vague question like "How can we improve our menu?" might be refined to "Which menu items are associated with the highest customer retention rates over a 6-month period?"

### 2. Determine Data Sources

Once you've defined your questions, the next step is to identify the data sources that will help you answer them. This involves:

- Conducting an inventory of all potential data sources within the organization, including **transactional databases**, **customer relationship management (CRM) systems**, and **external data sources**.
- Assess data quality to Evaluate the completeness, accuracy, and timeliness of each data source.
- Identify data gaps, and determine if additional data collection is necessary to answer your questions.
- Assess the feasibility of combining data from multiple sources.

This is an excellent time to revisit concepts like **data lakes**, which can store vast amounts of raw data in its native format, and **data mining/cleaning techniques**, which will be crucial for preparing your data for analysis. We've covered these in earlier chapters.

In the case of Juicy Lucy's, you might need to combine data from their point-of-sale systems, customer loyalty program, and even their "Rate-Your-Burp" app to get a comprehensive view of customer behavior and preferences.

### 3. Scoping and Gap Analysis

The final step in this process is to scope your analysis and identify any gaps between what you want to know and what your available data can tell you. This includes:

- Defining the **analysis scope**. This means clearly outlining what will and won't be included in your analysis.
- Identifying any constraints or limitations in your data or analytical capabilities.
- Performing a **gap analysis***--compare your ideal analysis scenario with what's actually possible given your current resources and data.
- Develop strategies to address any identified gaps, such as collecting additional data or employing advanced analytical techniques.

For Juicy Lucy's, you might realize that while you have extensive sales data, you lack information on customer sentiment beyond the "Rate-Your-Burp" app. This could lead to a proposal for implementing a more comprehensive customer feedback system.

By following this process, you'll be well-equipped to choose the most appropriate type of analysis for your specific business questions. Whether you end up performing trend analysis, exploratory data analysis, or any other technique, you'll have a solid foundation to ensure your analysis is focused, relevant, and actionable.

## How Can We Identify and Analyze Trends in Our Data?

**Trend analysis** is a crucial technique in data science that helps businesses understand patterns and changes in their data over time. For Juicy Lucy's Gourmet Burger Emporium, trend analysis can provide valuable insights into sales patterns, customer behavior, and the effectiveness of promotions.

Let's walk through the trend analysis process using a real example from Juicy Lucy's.

### Step 1: Define the metric and time frame

Clearly specify what you want to measure and over what period. This step is crucial as it sets the foundation for your analysis.

Example: Let's analyze monthly sales of Juicy Lucy's signature "Minnesotan Madness" burger over the past year.

### Step 2: Collect and prepare the data

Gather the relevant data from your data warehouse or other sources. This often involves writing SQL queries or using data extraction tools. Once collected, clean the data by handling missing values, removing duplicates, and ensuring consistent formatting.

Example: Let's create a sample dataset representing our monthly sales data:

```python
import pandas as pd
import numpy as np

# set seed
np.random.seed(52179)

# Create date range for the past three years (36 months)
date_range = pd.date_range(end=pd.Timestamp.now(), periods=36, freq='M')

# Generate sample sales data with a general upward trend and some seasonality
base_sales = 1000 + np.arange(36) * 50  # Upward trend
seasonality = np.sin(np.arange(36) * 2 * np.pi / 12) * 200  # Seasonal component
noise = np.random.normal(0, 50, 36)  # Random noise

sales = base_sales + seasonality + noise

# Create the DataFrame
df_sales = pd.DataFrame({
    'Date': date_range,
    'Sales': sales.astype(int)
})

df_sales.head()
```

This generates a DataFrame with monthly sales data, incorporating a general upward trend, seasonal fluctuations, and some random variation.

### Step 3: Visualize the data

Create visual representations of your data to get an initial understanding of trends. Time series plots are particularly useful for trend analysis.

Example: Let's create a line plot to visualize our sales data:

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
sns.lineplot(data=df_sales, x='Date', y='Sales', marker='o')
plt.title("Monthly Sales of 'Minnesotan Madness' Burger")
plt.xlabel("Date")
plt.ylabel("Sales (units)")
plt.grid(True)
plt.show()
```

This Seaborn lineplot provides a clear visualization of our sales data over time. The markers at each data point help to distinguish individual months, while the line connects the points to show the overall trend.

### Step 4: Identify patterns

Examine your visualizations and data to identify consistent upward or downward movements, cyclical patterns, or seasonal variations. Look for any sudden changes or anomalies that might require further investigation.

Example: From our plot, we can observe:

-   A general upward trend in sales over the year
-   Noticeable fluctuations, with peaks occurring around the summer months
-   No significant anomalies or sudden changes

### Step 5: Apply statistical techniques

Use statistical methods to quantify and confirm the trends you've observed visually. Common techniques include moving averages, exponential smoothing, and regression analysis.

Example: Let's run a linear regression:

```python
# Perform linear regression using statsmodels' formula API
import statsmodels.formula.api as smf
df_sales['Time'] = np.arange(len(df_sales))  # Creating a time variable
model = smf.ols('Sales ~ Time', data=df_sales).fit()

print(model.summary())
df_sales['Trend'] = model.fittedvalues
```

This output shows the results of an Ordinary Least Squares (OLS) regression analysis on 36 months of sales data. Let's break down the key points:

1. The R-squared value of 0.921 indicates that our model explains 91.8% of the variance in sales. This is a very good fit, suggesting that time is a strong predictor of sales.
2.  The F-statistic (398.1) and its p-value (2.40e-20) indicate that our model is statistically significant. The extremely low p-value suggests it's highly unlikely these results occurred by chance.
3.  The 'Time' coefficient (47) represents the average increase in sales per time period (likely month). This positive value indicates an upward trend in sales.
4.  The intercept (1062) represents the estimated sales at the beginning of our time series (time = 0).
5.  The [0.025 0.975] column shows the 95% confidence intervals for our coefficients. For both the intercept and time coefficient, these intervals don't include zero, further confirming their statistical significance.


In summary, this analysis reveals a strong, statistically significant upward trend in sales over the 36-month period. On average, sales increased by about 47 units per month. However, the low Durbin-Watson statistic suggests there might be additional time-dependent factors affecting sales that aren't captured by this simple linear trend.

Now, let's create a plot of this the "trend line" of this regression alongside the actual sales data a smoothed "3-month moving average)

```python
# Calculate 3-month moving average
df_sales['3_Month_MA'] = df_sales['Sales'].rolling(window=3).mean()

plt.figure(figsize=(12, 6))
sns.lineplot(data=df_sales, x='Date', y='Sales', marker='o', label='Actual Sales')
sns.lineplot(data=df_sales, x='Date', y='3_Month_MA', color='red', label='3-Month Moving Average')
sns.lineplot(data=df_sales, x='Date', y='Trend', color='green', label='Linear Trend')
plt.title("Monthly Sales Analysis")
plt.xlabel("Date")
plt.ylabel("Sales (units)")
plt.legend()
plt.grid(True)
plt.show()
```

This plot combines our original data, the 3-month moving average, and the linear trend line. The moving average smooths out short-term fluctuations, while the trend line shows the overall direction of sales over time.

### Step 6: Interpret results

Analyze your findings in the context of the business and external factors. Consider how internal decisions or external events might have influenced the observed trends.

Example: Interpreting our results:

-   The "Minnesotan Madness" burger shows a clear upward trend in sales over the year.
-   There's a noticeable seasonal pattern, with sales peaking in the summer months.
-   The 3-month moving average confirms the overall growth while capturing seasonal variations.
-   The linear trend line quantifies the average growth rate over time.

### Step 7: Make predictions

Based on the identified trends, forecast future values if appropriate. This could involve simple extrapolation or more complex time series forecasting
methods.

Example: While we won't perform a detailed forecast here, we could use our linear trend to make a simple projection.


```python
# use model to make prediction
df_sales['Forecast'] = model.predict(df_sales['Time'])
```

## How Can We Evaluate and Compare Performance Across Different Metrics?

**Performance analysis** is a critical technique in data science that allows businesses to assess how well they're meeting their goals and objectives. For Juicy Lucy's Gourmet Burger Emporium, performance analysis can help identify top-performing products, evaluate restaurant efficiency, and pinpoint areas for improvement.

Let's walk through the performance analysis process using examples from Juicy Lucy's.

### Step 1: Define Key Performance Indicators (KPIs)

Identify the specific metrics (**key performance indicators**) that best represent success for your business. These KPIs should align with your overall business objectives.

Example: For Juicy Lucy's, let's focus on three KPIs:
1. Average daily sales per restaurant
2. Customer satisfaction rating
3. Food waste percentage

### Step 2: Collect and prepare the data

Gather the relevant data for your chosen KPIs. This may involve querying multiple tables in your data warehouse and joining the results.

Example: Let's create a sample dataset with these KPIs for Juicy Lucy's restaurants:


```python
# Create sample data
np.random.seed(52179)
restaurants = ['Minneapolis', 'St. Paul', 'Duluth', 'Rochester', 'Bloomington']
dates = pd.date_range(start='2023-01-01', periods=30, freq='D')

# Define biases for each restaurant
biases = {
    'Minneapolis': {'sales_mean': 6000, 'sales_std': 800, 'satisfaction_min': 4, 'satisfaction_max': 5, 'waste_min': 3, 'waste_max': 8},
    'St. Paul': {'sales_mean': 5000, 'sales_std': 1000, 'satisfaction_min': 4.5, 'satisfaction_max': 5, 'waste_min': 4, 'waste_max': 9},
    'Duluth': {'sales_mean': 4500, 'sales_std': 900, 'satisfaction_min': 3.5, 'satisfaction_max': 4.5, 'waste_min': 2, 'waste_max': 6},
    'Rochester': {'sales_mean': 5200, 'sales_std': 950, 'satisfaction_min': 3.8, 'satisfaction_max': 4.7, 'waste_min': 5, 'waste_max': 10},
    'Bloomington': {'sales_mean': 4800, 'sales_std': 850, 'satisfaction_min': 4, 'satisfaction_max': 4.8, 'waste_min': 3, 'waste_max': 7},
}

data = {
    'Restaurant': [],
    'Date': [],
    'Avg_Daily_Sales': [],
    'Customer_Satisfaction': [],
    'Food_Waste_Pct': []
}

# Generate data
for date in dates:
    for restaurant in restaurants:
        params = biases[restaurant]
        data['Restaurant'].append(restaurant)
        data['Date'].append(date)
        data['Avg_Daily_Sales'].append(np.random.normal(params['sales_mean'], params['sales_std']))
        data['Customer_Satisfaction'].append(np.random.uniform(params['satisfaction_min'], params['satisfaction_max']))
        data['Food_Waste_Pct'].append(np.random.uniform(params['waste_min'], params['waste_max']))

df_performance = pd.DataFrame(data)
df_performance['Avg_Daily_Sales'] = df_performance['Avg_Daily_Sales'].astype(int)
df_performance['Customer_Satisfaction'] = df_performance['Customer_Satisfaction'].round(2)
df_performance['Food_Waste_Pct'] = df_performance['Food_Waste_Pct'].round(2)

df_performance.head()
```

This creates a DataFrame with performance data for five Juicy Lucy's restaurants over a 15-day period.

### Step 3: Calculate summary statistics
Compute relevant summary statistics to get an overall view of performance across your KPIs.

Example: Let's calculate mean performance for each restaurant:

```python
summary_stats = df_performance.groupby('Restaurant').agg({
    'Avg_Daily_Sales': 'mean',
    'Customer_Satisfaction': 'mean',
    'Food_Waste_Pct': 'mean'
}).round(2)

summary_stats
```

This gives us a quick overview of how each restaurant is performing across our KPIs.

### Step 4: Visualize the data

Create visual representations to help identify patterns, trends, and outliers in your performance data.

Example: Let's create a multi-metric visualization for our restaurants. In order to make this work, we'll need to **normalize** them (so that all the metrics are on a similar scale.

```python
# Normalize the metrics
df_normalized = df_performance.copy()
df_normalized['Avg_Daily_Sales'] = (df_normalized['Avg_Daily_Sales'] - df_normalized['Avg_Daily_Sales'].min()) / (df_normalized['Avg_Daily_Sales'].max() - df_normalized['Avg_Daily_Sales'].min())
df_normalized['Customer_Satisfaction'] = (df_normalized['Customer_Satisfaction'] - df_normalized['Customer_Satisfaction'].min()) / (df_normalized['Customer_Satisfaction'].max() - df_normalized['Customer_Satisfaction'].min())
df_normalized['Food_Waste_Pct'] = (df_normalized['Food_Waste_Pct'] - df_normalized['Food_Waste_Pct'].min()) / (df_normalized['Food_Waste_Pct'].max() - df_normalized['Food_Waste_Pct'].min())

# drop date
df_normalized = df_normalized.drop('Date', axis=1)

# Compute summary statistics excluding the 'Date' column
summary_stats_norm = df_normalized.groupby('Restaurant').mean().reset_index()

# Melt the DataFrame to long format for easier plotting
df_melted = pd.melt(summary_stats_norm, id_vars=['Restaurant'], var_name='Metric', value_name='Value')

df_melted.head()
```

### Bar Graph: Normalized Key Performance Indicators

```python
# Create the plot
plt.figure(figsize=(12, 6))
sns.barplot(data=df_melted, x='Restaurant', y='Value', hue='Metric')
plt.title("Performance Metrics by Restaurant")
plt.xticks(rotation=45)
plt.legend(title='Metric', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

This plot allows us to compare all three metrics across restaurants simultaneously.

### Step 5: Identify top performers and areas for improvement

Analyze your visualizations and summary statistics to determine which units are performing well and which may need attention.

Example: From our summary and visualization, we can observe:

-   Minneapolis seems to have the highest average daily sales
-   St. Paul has the highest customer satisfaction rating
-   Duluth has the lowest food waste percentage

### Step 6: Perform comparative analysis

Compare performance across different dimensions (e.g., time, location) to gain deeper insights.

Example: Let's compare daily sales performance over time for our top and bottom performing restaurants:

```python
top_restaurant = summary_stats['Avg_Daily_Sales'].idxmax()
bottom_restaurant = summary_stats['Avg_Daily_Sales'].idxmin()

plt.figure(figsize=(12, 6))
for restaurant in [top_restaurant, bottom_restaurant]:
    data = df_performance[df_performance['Restaurant'] == restaurant]
    sns.lineplot(data=data, x='Date', y='Avg_Daily_Sales', label=restaurant)

plt.title("Daily Sales Comparison: Top vs Bottom Performing Restaurants")
plt.xlabel("Date")
plt.ylabel("Average Daily Sales ($)")
plt.legend(title='Restaurant')
plt.grid(True)
plt.show()
```

This allows us to see how the top and bottom performers compare over time, potentially revealing patterns or anomalies.
